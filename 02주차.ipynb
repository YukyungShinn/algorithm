```python
n = int(input())
temp = n
count = 0

while True :
	left = temp // 10
	right = temp % 10
	add = (left + right) % 10
	temp = (right * 10) + add

	count += 1
	if temp == n:
		print(count)
		break
```

```python
n=int(input())
grade=list()
g2=list()
grade.extend(map(int,input().split()))

maxgrade=max(grade)
for i in range(n):
    g2.append(grade[i]/maxgrade*100)

sumgrade=sum(g2)
result=sumgrade/len(g2)

print(result)
```

```python
n = int(input())

for _ in range(n):
    a = list(map(int, input().split()))
    mean = sum(a[1:])/a[0]
    b = 0
    for i in a[1:]:
        if i > mean:
            b += 1
    perc = b/a[0] *100
    print(f"{perc:.3f}%")
```
